### Filtrando os dados na planilha Vivara

```python
import pandas as pd
import matplotlib.pyplot as plt
import mplcyberpunk

plt.style.use('cyberpunk')
```

```python
vivara_planilha = pd.read_excel('Dados Históricos  2T24.xlsx', sheet_name='Lojas | Stores', skiprows=3, usecols = ['Data de Abertura', 'Estado', 'Cidade', 'Localização', 'Tipo', 'Status'])

vivara_planilha
```

```python
vivara_planilha = vivara_planilha.dropna()

vivara_planilha
```

```python
vivara_planilha = vivara_planilha[(vivara_planilha['Tipo'] == 'Loja Vivara') & (vivara_planilha['Status'] == 'Aberto')]

vivara_planilha
```

```python
vivara_planilha['Data de Abertura'] = pd.to_datetime(vivara_planilha['Data de Abertura']).dt.date

vivara_planilha
```

### Pegar os dados dos municípios

```python
pib_municipios = pd.read_excel('pib_municipios.xlsx')
pop_municipios = pd.read_excel('populacao_municipios.xlsx')

pop_municipios
```

```python
pib_municipios
```

```python
dados_municipios = pib_municipios.merge(pop_municipios, on = "City")

dados_municipios
```

```python
# FAzer o PIB per capta

dados_municipios['pib_per_capta'] = dados_municipios['pib'] / dados_municipios['pop']

dados_municipios
```

```python
dados_municipios[['City', 'Estado']] = dados_municipios['City'].str.split('(', expand = True)

dados_municipios
```

```python
dados_municipios['Estado'] = dados_municipios['Estado'].str.replace(')', '')

dados_municipios
```

```python
dados_municipios.to_excel("dados_municipios.xlsx")
```

```python
#Renomear nome da coluna Cidade pra facilitar o join

vivara_planilha = vivara_planilha.rename(columns = {'Cidade': 'City'})

vivara_planilha
```

### Juntar os dados da Planilha da Vivara com o do IBGE

```python
dados_municipios['City'] = dados_municipios['City'].str.strip()
dados_municipios['Estado'] = dados_municipios['Estado'].str.strip()
vivara_planilha['Estado'] = vivara_planilha['Estado'].str.strip()
vivara_planilha['City'] = vivara_planilha['City'].str.strip()

dados_municipios
```

```python
# Fazendo o merge com 'outer' para ver o que não deu match
dados_lojas_e_economia = dados_municipios.merge(vivara_planilha, on=['City', 'Estado'], how='outer', indicator=True)

# Visualizando onde o merge falhou
cidades_nao_encontradas = dados_lojas_e_economia[dados_lojas_e_economia['_merge'] != 'both']

# Exibir o resultado
cidades_nao_encontradas[['City', 'Estado', '_merge']]
```

```python
# Identificar a cidade que está na planilha e não bateu com a do IBGE - Ajustar a planilha da Vivara se aparecer alguma cidade

# Erro de digitação no da Vivara

dados_lojas_e_economia_both = dados_lojas_e_economia[dados_lojas_e_economia['_merge'] == 'right_only']

# Exibir o resultado
dados_lojas_e_economia_both
```

```python
#Aqui to dando merge nas duas planilhas pra deixar tudo em um só e que só bate as cidades que tem o mesmo nome e estado também

dados_lojas_e_economia = dados_municipios.merge(vivara_planilha, on = ['City', 'Estado'])

dados_lojas_e_economia
```
